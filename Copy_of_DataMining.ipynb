## Normalization

```python

from sklearn.preprocessing import MinMaxScaler

Scaler=MinMaxScaler()

Sales=data1['Sales'].values.reshape(-1,1)
Profit=data1['Profit'].values.reshape(-1,1)

normalize_Sales= Scaler.fit_transform(Sales)
normalize_Profit= Scaler.fit_transform(Profit)

data1['Sales'] = normalize_Sales
data1['Profit'] = normalize_Profit
```

```python
UnitPrice=data1['Unit Price'].values.reshape(-1,1)
ShippingCost=data1['Shipping Cost'].values.reshape(-1,1)

normalize_UnitPrice = Scaler.fit_transform(UnitPrice)
normalize_ShippingCost= Scaler.fit_transform(ShippingCost)

data1['UnitPrice'] = normalize_UnitPrice
data1['ShippingCost'] = normalize_ShippingCost

data1.head(5)
```

## Product Category Classification using logistic Regression

```python
x=data1[['Product Container','Unit Price']]
y=data1[['Product Category']]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

x_train , x_test , y_train , y_test =train_test_split(x,y, test_size=0.30)

scaler=StandardScaler()
scaled_x_train=scaler.fit_transform(x_train)
scaled_x_test =scaler.fit_transform(x_test)
```

```python
from sklearn.metrics import accuracy_score
def modelresults(predictions):
  print("accuracy score is :{} ".format(accuracy_score(y_test,predictions)))
```

```python
from sklearn.linear_model import LogisticRegression
log_model=LogisticRegression()
log_model.fit(scaled_x_train,y_train)
```

```python
log_prediciton=log_model.predict(scaled_x_test)
modelresults(log_prediciton)
```

```python
import numpy as np
newProduct=np.array([5,0.084])
```

```python
#from sklearn.metrics._plot.regression import PredictionErrorDisplay
def Product(newProduct):
    pr = log_model.predict(newProduct.reshape(1,-1))
    if pr==1:
        return "offices supplies"
    elif pr==0:
         return "furniture"
    else:
      return "technology"
```

```python
print("this product category is : {}".format(Product(newProduct)))
```
